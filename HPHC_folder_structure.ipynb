# Hospital Patient Helpdesk Chatbot - Project Scaffold

This notebook creates a clean, sequentially numbered folder structure for an end-to-end healthcare support RAG chatbot. It separates ingestion, embeddings, retrieval, safety, API, UI, evaluation, monitoring, deployment, and tests.

> The chatbot must provide administrative and informational support only. It must not diagnose conditions, prescribe treatment, or recommend medication dosages. Emergency messages must direct users to local emergency services and qualified medical professionals.

```python
def python_stub(path: str, purpose: str) -> str:
    module_name = Path(path).stem
    return dedent(
        f'''\
        """{purpose}"""

        from __future__ import annotations


        def main() -> None:
            """Run the {module_name} workflow."""
            raise NotImplementedError("Implement {module_name} in the recommended build order.")


        if __name__ == "__main__":
            main()
        '''
    )


def test_stub(purpose: str) -> str:
    return dedent(
        f'''\
        """{purpose}"""

        import pytest


        @pytest.mark.skip(reason="Enable after the corresponding module is implemented.")
        def test_placeholder() -> None:
            assert True
        '''
    )


def notebook_stub(filename: str, purpose: str) -> str:
    """Build a valid starter notebook for one workflow stage."""
    title = Path(filename).stem.replace("_", " ").title()
    function_name = Path(filename).stem.split("_", maxsplit=1)[-1]
    notebook = {
        "cells": [
            {
                "cell_type": "markdown",
                "id": "overview",
                "metadata": {},
                "source": [
                    f"# {title}\n",
                    "\n",
                    f"**Purpose:** {purpose}\n",
                    "\n",
                    "This notebook is one stage of the Hospital Patient Helpdesk Chatbot RAG workflow."
                ],
            },
            {
                "cell_type": "markdown",
                "id": "workflow",
                "metadata": {},
                "source": [
                    "## Workflow\n",
                    "1. Load configuration and validate inputs.\n",
                    "2. Run this stage with small, approved sample data.\n",
                    "3. Inspect outputs and record assumptions.\n",
                    "4. Add tests before promoting logic to the matching Python module."
                ],
            },
            {
                "cell_type": "code",
                "execution_count": None,
                "id": "setup",
                "metadata": {},
                "outputs": [],
                "source": [
                    "from __future__ import annotations\n",
                    "\n",
                    "from pathlib import Path\n",
                    "\n",
                    "PROJECT_ROOT = Path.cwd().resolve().parent\n",
                    "PROJECT_ROOT\n",
                ],
            },
            {
                "cell_type": "code",
                "execution_count": None,
                "id": "starter-function",
                "metadata": {},
                "outputs": [],
                "source": [
                    f"def {function_name}() -> None:\n",
                    f"    \"\"\"{purpose}\"\"\"\n",
                    "    raise NotImplementedError(\"Implement and verify this workflow stage.\")\n",
                ],
            },
            {
                "cell_type": "markdown",
                "id": "safety-check",
                "metadata": {},
                "source": [
                    "## Safety and quality check\n",
                    "- Use approved or synthetic hospital data.\n",
                    "- Do not expose personal or protected health information.\n",
                    "- Keep generated answers grounded in cited hospital sources.\n",
                    "- Do not provide diagnosis, treatment, or dosage recommendations."
                ],
            },
        ],
        "metadata": {
            "kernelspec": {
                "display_name": "Python 3",
                "language": "python",
                "name": "python3",
            },
            "language_info": {"name": "python", "version": "3.12"},
        },
        "nbformat": 4,
        "nbformat_minor": 5,
    }
    return json.dumps(notebook, indent=1) + "\n"


FILE_SPECS = [FileSpec(path, purpose, python_stub(path, purpose)) for path, purpose in PYTHON_MODULES.items()]
FILE_SPECS += [FileSpec(path, purpose, test_stub(purpose)) for path, purpose in TEST_MODULES.items()]
FILE_SPECS += [FileSpec(path, "Project data placeholder.", content) for path, content in DATA_FILES.items()]
FILE_SPECS += [
    FileSpec(
        folder_path("notebooks", f"{Path(path).stem}.ipynb"),
        purpose,
        notebook_stub(Path(path).name, purpose),
    )
    for path, purpose in PYTHON_MODULES.items()
]
```

## Create the project
Existing files are preserved by default. Legacy unnumbered folders are renamed to the new `01_` through `13_` convention. Set `OVERWRITE_EXISTING = True` only when replacement is intentional.

```python
def write_text_file(spec: FileSpec, *, overwrite: bool) -> str:
    destination = PROJECT_ROOT / spec.path
    destination.parent.mkdir(parents=True, exist_ok=True)

    if destination.exists() and not overwrite:
        return "preserved"

    status = "updated" if destination.exists() else "created"
    destination.write_text(spec.content or "", encoding="utf-8")
    return status


def migrate_legacy_folders() -> int:
    """Rename folders created by the earlier unnumbered scaffold."""
    migrated_count = 0
    for legacy_name, numbered_name in FOLDERS.items():
        legacy_path = PROJECT_ROOT / legacy_name
        numbered_path = PROJECT_ROOT / numbered_name

        if not legacy_path.exists():
            continue
        if numbered_path.exists():
            raise FileExistsError(
                f"Cannot migrate {legacy_path}: destination already exists: {numbered_path}"
            )

        legacy_path.rename(numbered_path)
        migrated_count += 1

    return migrated_count


def remove_legacy_notebook() -> bool:
    """Remove the obsolete empty notebook from the earlier scaffold."""
    legacy_notebook = PROJECT_ROOT / folder_path("notebooks", "chatbot_experiment.ipynb")
    if not legacy_notebook.exists():
        return False

    legacy_notebook.unlink()
    return True


def create_project(*, overwrite: bool = False) -> dict[str, int]:
    results = {"created": 0, "updated": 0, "preserved": 0, "migrated": 0, "removed": 0}
    PROJECT_ROOT.mkdir(parents=True, exist_ok=True)
    results["migrated"] = migrate_legacy_folders()
    results["removed"] = int(remove_legacy_notebook())

    for spec in FILE_SPECS:
        is_workflow_notebook = (
            Path(spec.path).parent.name == FOLDERS["notebooks"]
            and Path(spec.path).suffix == ".ipynb"
        )
        status = write_text_file(
            spec,
            overwrite=overwrite or (is_workflow_notebook and REFRESH_WORKFLOW_NOTEBOOKS),
        )
        results[status] += 1

    logo_path = PROJECT_ROOT / folder_path("app", "assets/hospital_logo.png")
    logo_path.parent.mkdir(parents=True, exist_ok=True)
    if overwrite or not logo_path.exists():
        logo_path.write_bytes(base64.b64decode(LOGO_PNG_BASE64))
        results["created"] += 1
    else:
        results["preserved"] += 1

    return results


creation_summary = create_project(overwrite=OVERWRITE_EXISTING)
print(f"Project root: {PROJECT_ROOT}")
print(
    f"Created: {creation_summary['created']} | "
    f"Updated: {creation_summary['updated']} | "
    f"Preserved: {creation_summary['preserved']} | "
    f"Migrated folders: {creation_summary['migrated']} | "
    f"Removed legacy files: {creation_summary['removed']}"
)
```

```python
def render_tree(root: Path) -> str:
    lines = [f"{root.name}/"]

    def add_directory(directory: Path, indent: str = "") -> None:
        entries = sorted(directory.iterdir(), key=lambda item: (not item.is_dir(), item.name.lower()))
        for entry in entries:
            suffix = "/" if entry.is_dir() else ""
            lines.append(f"{indent}+-- {entry.name}{suffix}")
            if entry.is_dir():
                add_directory(entry, indent + "    ")

    add_directory(root)
    return "\n".join(lines)


print(render_tree(PROJECT_ROOT))
```

## Validate the result
The final check fails loudly when a requested file is missing, making the notebook suitable for repeatable project setup.

```python
def validate_project() -> None:
    expected_paths = [PROJECT_ROOT / spec.path for spec in FILE_SPECS]
    expected_paths.append(PROJECT_ROOT / folder_path("app", "assets/hospital_logo.png"))
    missing_paths = [path for path in expected_paths if not path.exists()]

    if missing_paths:
        formatted = "\n".join(f"- {path}" for path in missing_paths)
        raise FileNotFoundError(f"Project scaffold is incomplete:\n{formatted}")

    notebooks_directory = PROJECT_ROOT / FOLDERS["notebooks"]
    expected_notebooks = {f"{Path(path).stem}.ipynb" for path in PYTHON_MODULES}
    actual_notebooks = {path.name for path in notebooks_directory.glob("*.ipynb")}
    if actual_notebooks != expected_notebooks:
        raise ValueError(
            f"Unexpected notebook set. Expected {sorted(expected_notebooks)}, "
            f"found {sorted(actual_notebooks)}"
        )

    for notebook_path in notebooks_directory.glob("*.ipynb"):
        notebook = json.loads(notebook_path.read_text(encoding="utf-8"))
        if notebook.get("nbformat") != 4 or not notebook.get("cells"):
            raise ValueError(f"Invalid starter notebook: {notebook_path}")

    print(
        f"Validation passed: {len(expected_paths)} expected files and "
        f"{len(expected_notebooks)} numbered notebooks are present."
    )


validate_project()
```
